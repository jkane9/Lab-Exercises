```python
from tkinter import *
import requests
import json
import random

root = Tk()
root.title("Trivia Game")
root.geometry("1000x800")

canvas = Canvas(root, width=800, height=250, bg="light blue")
canvas.place(relx=.5, rely=.5,anchor= CENTER)

game_display = canvas.create_text(200, 75, text="", fill="black", font=('Helvetica 15 bold'))

feedback = Label(root, text="", pady=10, font=("arial", 15, "bold"))

user_answer = StringVar()
user_answer.set("")

score = 0
questions_correct = 0

trivia_data = requests.get(f"https://the-trivia-api.com/api/questions?categories=film_and_tv&limit=10&region=US&difficulty=easy")
trivia_json = trivia_data.json()

def questions(json):
    trivia_questions = []
    for i in trivia_json:
        question = i['question']
        trivia_questions.append(question)
    return trivia_questions

def random_question(ques_list):
    rand_ques = random.choice(ques_list)
    return rand_ques

def answers(json, individual_ques):
    answers1 = []
    for i in trivia_json:
        if i['question'] == individual_ques:
            cor_answer = i["correctAnswer"]
            answers1.append(cor_answer)
            for k in i["incorrectAnswers"]:
                answers1.append(k)
    random.shuffle(answers1)
    return answers1

def ask_question(individual_ques, answers2):
    canvas.itemconfig(game_display, text=individual_ques, width=300)
    
def gen_radiobuttons(answers2):
    y_pos = 100
    for i in answers2:
        rad_btn = Radiobutton(canvas, text=i, variable=user_answer, command=eval_score, value=i, font=('Helvetica 10 bold'))
        rad_btn.place(x=50, y=y_pos)
        y_pos+=20
        
def eval_score():
    answer = user_answer.get()
    for i in trivia_json:
        if i["question"] == current_question:
            if i["correctAnswer"] == answer:
                feedback["text"] = "Correct!"
                feedback.place(x=400, y=500)
                global score
                global questions_correct
                score += 10
                questions_correct += 1
                score_display.config(text=f"Your score is {str(score)}")
                correct_display.config(text=f"You have gotten {str(questions_correct)} questions correct!")
            else:
                feedback["text"] = "Incorrect."
                feedback.place(x=400, y=500)
                                
def main(): 
    q_list = questions(trivia_json)
    global current_question
    current_question = random_question(q_list)
    ans_list = answers(trivia_json, current_question)
    ask_question(current_question, ans_list)
    gen_radiobuttons(ans_list)
    
new_button = Button(root, text = "Click here to get a new question", command = main)
score_display = Label(root, text = f"Your score is {str(score)}")
correct_display = Label(root, text = f"You have gotten {str(questions_correct)} questions correct!")
    
main()
new_button.pack()
score_display.pack()
correct_display.pack()
root.mainloop()

#I was able to create a button to grab a new question and its' corresponding answers, however a problem I ran into
#was that the button would lead to some repeated questions, and also did not fully delete the previous answers. This
#would lead to those previous answers sometimes appearing underneath the current answers. I tried to create a function 
#to combat this, however I could not get around a way to delete the current answers before giving the user a chance to 
#answer the question (as the button would have to rely on the main() function). However, it still keeps score and track of correctly
#answered questions while clicking through new questions.

#For creating the score counter, I refered to Kathie Quick's youtube video which helped me realize that I needed to
#configure the score display within the eval_score function as well. The link to her video is:
# https://www.youtube.com/watch?v=TosEDC0Y7E4

#Probably relying a bit on the new question button, I was also unable to fully figure out a way to display the correctly answered
#questions AFTER all the questions had been run through. Again I tried a function, however because my new question 
#button would sometimes show repeated questions/answers, it would also not stop after all questions had been run through, so 
#I did not have a way to display the number of correct answers after they were all asked. However, like score, I was able
#to display the number of correct answers throughout new questions.
```
