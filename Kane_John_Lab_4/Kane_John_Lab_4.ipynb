# IMDB Webscrape

```python
from bs4 import BeautifulSoup
import requests
import csv
import html5lib
```

```python
site = requests.get("https://www.imdb.com/list/ls055592025/").text
soup = BeautifulSoup(site, "lxml")
```

```python
info = soup.find("div", class_ = "lister-item-content")
print(info.prettify())
```

```python
title = info.find("a").text
print(f"Title: {title}")

year = info.find("span", class_ = "lister-item-year text-muted unbold").text
print(f"Release Year: {year}")

imdb_rating = info.find("span", class_ = "ipl-rating-star__rating").text
print(f"IMDB Rating: {imdb_rating}/10")
```

```python
csv_file = open("imdb_web_scrape.csv", "w", newline = "", encoding = "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Title", "Year","IMDB Rating"])

for movie in soup.find_all("div", class_ = "lister-item-content"):
    title = movie.find("a").text
    print(f"Title: {title}")

    year = movie.find("span", class_ = "lister-item-year text-muted unbold").text
    print(f"Release Year: {year}")

    imdb_rating = movie.find("span", class_ = "ipl-rating-star__rating").text
    print(f"IMDB Rating: {imdb_rating}/10")
    
    print("...")
    
    csv_writer.writerow([title,year,imdb_rating])
    
csv_file.close()
```

# Library of Congress

```python
site2 = requests.get("https://www.loc.gov/search/?q=cats&sp=1").text
soup2 = BeautifulSoup(site2, "lxml")
```

```python
info2 = soup2.find("div", class_ = "description")
print(info2.prettify())
```

```python
name = info2.find("a").text.strip()
print(f"Name: {name}")

description = info2.find("span", class_ = "original-format").text.strip()
print(f"Description: {description}")

hyperlink = info2.find("a").get("href")
print(f"Hyperlink: {hyperlink}")
```

```python
for item in soup2.find_all("div", class_ = "description"):
    name = item.find("a").text.strip()
    print(f"Name: {name}")

    description = item.find("span", class_ = "original-format").text.strip()
    print(f"Description: {description}")

    hyperlink = item.find("a").get("href")
    print(f"Hyperlink: {hyperlink}")
    
    print("...")
```

```python
csv_file2 = open("congress_cats.csv", "w", newline = "", encoding = "utf-8")
csv_writer = csv.writer(csv_file2)
csv_writer.writerow(["Name", "Description","Hyperlink"])

page = 1

while page != 6:
    cats = requests.get(f"https://www.loc.gov/search/?q=cats&sp={page}").text
    soup2 = BeautifulSoup(cats, "lxml")
    
    for item in soup2.find_all("div", class_ = "description"):
        name = item.find("a").text.strip()
        print(f"Name: {name}")

        description = item.find("span", class_ = "original-format").text.strip()
        print(f"Description: {description}")

        
        hyperlink = item.find("a").get("href")
        print(f"Hyperlink: {hyperlink}")

        print("...")
        
        csv_writer.writerow([name,description,hyperlink])

    
    page += 1
    
csv_file2.close()
```
