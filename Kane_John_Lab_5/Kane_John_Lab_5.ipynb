# Selenium Tutorial 3

```python
PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service = PATH)

driver.get("https://techwithtim.net")  #Restablishing the driver as the teachwithtim website

link = driver.find_element(By.LINK_TEXT, "Python Programming") #locates link tag
link.click() #this command clicks on that link and brings up the page

try:   #have to wait again in order to avoid a crash with the program running too fast
    element = WebDriverWait(driver, 10).until(  
    EC.presence_of_element_located((By.LINK_TEXT, "Beginner Python Tutorials")) #locates beginner python tutorial link on python programming page
    )
    element.click() #clicks the button to load the "Beginner Python Tutorials" page
    
    element = WebDriverWait(driver, 10).until(  #have to wait again in order to get the next page to load
    EC.presence_of_element_located((By.ID, "sow-button-19310003")) #locates the "Get Started" button on the beginner python tutorial page
    )
    element.click() #clicks the button to load the "Get Started" page
    
    driver.back() #brings the page back to previous page
    driver.back()
    driver.back() #doing this 3 times to go back to home page
    driver.forward()
    driver.forward() #brings you forward to a page you were just on, going back to "Beginner Python Tutorials" page
except:
    driver.quit()
    
#element.clear() clears the text already in the field, so that new text replaces it, and is not appended to it
```

# Climate Change

```python
PATH = Service("C:\Program Files (x86)\chromedriver.exe")
driver = webdriver.Chrome(service = PATH)

csv_file = open("climate_change.csv", "w", newline = "", encoding = "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Title", "Description"])

driver.get("https://data.gov/")

link = driver.find_element(By.CLASS_NAME, "menu-data") 
link.click()

search = driver.find_element(By.ID, "search-big")
search.clear()
search.send_keys("climate change")
search.send_keys(Keys.RETURN)


try:
    content = WebDriverWait(driver, 10).until(  
    EC.presence_of_element_located((By.CLASS_NAME, "module-content")) 
    )

    story = content.find_elements(By.CLASS_NAME, "dataset-content")
    
    for i in story:
        title = i.find_element(By.CLASS_NAME, "dataset-heading")
        description = i.find_element(By.CLASS_NAME, "notes")
        print(f"Title: {title.text}")
        print("...")
        print(f"Description: {description.text}")
        print("...")
        
        csv_writer.writerow([title.text,description.text])
                
finally:
    driver.quit()  
    
csv_file.close()
```
